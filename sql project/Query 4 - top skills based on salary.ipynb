## Overview

This visualization highlights the **top 10 highest-paying skills** for Remote Data Analysts, based on average annual salary data.

## Methodology

The analysis calculates the average salary for each skill mentioned in remote job postings, focusing exclusively on entries with valid salary information. The final chart was generated using the **Matplotlib** library in Python to compare these top-tier earners.

## Process Summary

* [x] Execute SQL query to identify the top 10 skills by average salary for 'Data Analyst' roles.
* [x] Filter for remote positions (`job_work_from_home = True`) and non-null salary data.
* [x] Group results by skill and sort by descending average salary to isolate the highest earners.
* [x] Export the SQL results as a JSON/DataFrame for visualization.
* [x] Use Python and Matplotlib to create a clear, formatted horizontal bar chart.

This approach ensures that the insights directly represent the most lucrative technical skills in the remote data job market, providing a clear financial roadmap for career development.

**Query 3:**
```sql

SELECT
    skills,
    ROUND(AVG(salary_year_avg), 0) AS avg_salary
FROM job_postings_fact
INNER JOIN skills_job_dim ON job_postings_fact.job_id = skills_job_dim.job_id
INNER JOIN skills_dim ON skills_job_dim.skill_id = skills_dim.skill_id
WHERE
    job_title_short = 'Data Analyst'
    AND salary_year_avg IS NOT NULL
    AND job_work_from_home = True
GROUP BY
    skills
ORDER BY
    avg_salary DESC
LIMIT 10;
```


````python
import matplotlib.pyplot as plt

# הנתונים מהתמונה שלך
skills = ['pyspark', 'bitbucket', 'watson', 'couchbase', 'datarobot', 'gitlab', 'swift', 'jupyter', 'pandas', 'elasticsearch']
salaries = [208172, 189155, 160515, 160515, 155486, 154500, 153750, 152777, 151821, 145000]

# יצירת הגרף (גרף עמודות אופקי נוח יותר לקריאת שמות הכישורים)
plt.barh(skills, salaries, color='skyblue')

# הוספת כותרות
plt.title('Top 10 High Paying Skills')
plt.xlabel('Average Salary ($)')
plt.ylabel('Skills')

# הפיכת הסדר כדי שהשכר הגבוה ביותר יהיה למעלה
plt.gca().invert_yaxis()

# הצגת הגרף
plt.tight_layout()
plt.show()
````
