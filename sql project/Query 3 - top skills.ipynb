This ensures that the chart directly reflects the SQL analysis results and provides actionable insights for aspiring data analysts.

**Query 3:**
```sql
WITH top_5_skills AS (
    SELECT
        skills,
        COUNT(*) as demand
    FROM
        job_postings_fact
    INNER JOIN skills_job_dim ON job_postings_fact.job_id = skills_job_dim.job_id
    INNER JOIN skills_dim ON skills_job_dim.skill_id = skills_dim.skill_id
    WHERE
        job_title_short = 'Data Analyst'
    GROUP BY
        skills
    ORDER BY
        demand DESC
    LIMIT 5
)
SELECT *
FROM top_5_skills;

```

````python
import matplotlib.pyplot as plt

# Data
skills = ['sql', 'excel', 'python', 'tableau', 'power bi']
demand = [92628, 67031, 57326, 46554, 39468]

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(demand, labels=skills, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 In-Demand Data Analyst Skills')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````
